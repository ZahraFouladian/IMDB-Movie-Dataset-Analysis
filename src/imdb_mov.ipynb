###  Import Libraries 

```python
# Importing the necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation and analysis
from skimpy import skim  # For quickly summarizing a DataFrame (similar to .describe)
import matplotlib.pyplot as plt  # For creating static visualizations (plots)
import seaborn as sns  # For creating more advanced visualizations
import plotly.express as px  # For interactive visualizations
```

### Loading & Reading Dataset

```python
# Reading the "imdb_movie_dataset.csv" file into a DataFrame, using the "Rank" column as the index
df = pd.read_csv("datasets/imdb_movie_dataset.csv", index_col="Rank")

# Displaying the DataFrame to check the content
df
```

### Top 10 Directors Who Made the Most Money

```python
# Grouping the DataFrame by 'Director' and calculating the total revenue for each director,
# sorting the results in descending order and selecting the top 10 highest-grossing directors
top_10_highest_grossing_directors = df.groupby("Director")['Revenue (Millions)'].sum().sort_values(ascending=False)[:10]

# Displaying the top 10 highest-grossing directors
top_10_highest_grossing_directors
```

```python
# Setting the style of the seaborn plot to 'whitegrid' for better visibility of the gridlines
sns.set_style("whitegrid")

# Creating a figure with specified size
plt.figure(figsize=(10, 6))

# Creating a bar plot for the top 10 highest-grossing directors with custom colors and removing the legend
ax = sns.barplot(x=top_10_highest_grossing_directors.values, y=top_10_highest_grossing_directors.index, hue=top_10_highest_grossing_directors.index, palette="turbo", legend=False)

# Adding labels on top of each bar in the plot to show the exact revenue values
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)

# Labeling the x-axis and y-axis
plt.xlabel('Revenue (Millions)', fontsize=12)
plt.ylabel('Directors', fontsize=12)

# Adding a bold title to the plot
plt.title("Top 10 Directors Who Made the Most Money", fontsize=14, fontweight='bold')

# Displaying the plot
plt.show()
```

### Top 10 Directors Who Directed the Most Movies

```python
# Counting the number of movies directed by each director, sorting in descending order, 
# and selecting the top 10 directors with the most movies
director_counts = df['Director'].value_counts().sort_values(ascending=False)[:10]

# Displaying the top 10 directors with the most movies
director_counts
```

```python
# Setting the style of the seaborn plot to 'whitegrid' for better visibility of the gridlines
sns.set_style("whitegrid")

# Creating a figure with a specified size
plt.figure(figsize=(10, 6))

# Labeling the x-axis and y-axis
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Directors', fontsize=12)

# Adding a bold title to the plot
plt.title("Top 10 Directors Who Directed the Most Movies", fontsize=14, fontweight='bold')

# Creating a bar plot to display the number of movies directed by the top 10 directors
ax = sns.barplot(x=director_counts.values, y=director_counts.index, hue=director_counts.index, palette="turbo", legend=False)

# Adding labels on top of each bar to show the exact count of movies directed
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)

# Displaying the plot
plt.show()
```

### The Directors Of The Top 10 Movies Based On Imdb Ratings

```python
# Selecting the top 10 movies with the highest ratings, displaying only the "Title", "Director", and "Rating" columns,
# and resetting the index of the resulting DataFrame
top_10_movies = df.nlargest(10, "Rating")[["Title", "Director", "Rating"]].reset_index(drop=True)

# Displaying the top 10 movies with the highest ratings
top_10_movies
```

```python
# Creating a figure with a specified size
plt.figure(figsize=(10, 6))

# Creating a horizontal bar plot for the top 10 movies by rating
# Using 'viridis' color palette and including movie titles as hue
sns.barplot(x='Rating', y='Title', data=top_10_movies, hue='Title', palette='viridis')

# Adding text annotations next to each bar to display the director of each movie
for i, row in top_10_movies.iterrows():
    plt.text(row['Rating'] + 0.05, i, row['Director'], fontsize=9, va='center')

# Adding a bold title to the plot
plt.title('Top 10 Movies by Rating', fontsize=14, fontweight='bold')

# Labeling the x-axis and y-axis
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)

# Setting the x-axis limits to focus on the top ratings
plt.xlim(8.4, 9.1)

# Displaying the plot
plt.show()
```

### Correlation Matrix of Rating, Votes, Revenue, and Metascore

```python
# Calculating the correlation matrix for the specified columns ('Rating', 'Votes', 'Metascore', and 'Revenue (Millions)') 
# to see how they are related to each other in terms of linear relationships.
correlation_matrix = df[['Rating', 'Votes', 'Metascore', 'Revenue (Millions)']].corr()

# Displaying the resulting correlation matrix
correlation_matrix
```

```python
# Setting up the figure size for the heatmap
plt.figure(figsize=(8, 6))

# Creating a heatmap to visualize the correlation matrix with annotations and the "Blues" color palette
# The fmt=".2f" ensures that the correlation values are shown with two decimal places
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", cbar=True)

# Adding a title to the heatmap
plt.title('Correlation Matrix of Rating, Votes, Revenue, and Metascore')

# Displaying the heatmap
plt.show()
```

Rating and Votes have a strong positive correlation (0.51), meaning when one increases, the other tends to increase too.

Rating and Metascore are somewhat related (0.63), suggesting they move together, but not as strongly.

Votes and Revenue have a high correlation (0.64), showing that more votes may lead to higher revenue.

Metascore and Revenue have a weak correlation (0.14), indicating not much relationship between them.

### Number of Movies Released per Year

```python
# Counting the number of movies released each year by using the 'Year' column and sorting them
movies_per_year = df['Year'].value_counts()

# Displaying the number of movies per year
movies_per_year
```

```python
# Creating a figure with a specified size for the line plot
plt.figure(figsize=(10, 6))

# Creating a line plot to show the number of movies released per year, with circular markers at each data point
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values, marker='o')

# Adding a bold title to the plot
plt.title('Number of Movies Released per Year', fontsize=12, fontweight='bold')

# Labeling the x-axis and y-axis
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)

# Enabling the grid for better readability of the plot
plt.grid(True)

# Displaying the plot
plt.show()
```

### Top 20 Actors/Actress Appeared In The Most Movies

```python
# Splitting the 'Actors' column by commas to create a list of actors for each movie
df['Actors_list'] = df['Actors'].str.split(",")

# Exploding the list of actors so that each actor is in a separate row, and stripping any extra spaces
df1 = df.explode("Actors_list")["Actors_list"].str.strip()

# Counting the occurrences of each actor and selecting the top 20 actors who appeared the most
top_actors = df1.value_counts().sort_values(ascending=False)[:20]

# Displaying the top 20 actors
top_actors
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_actors.values, y=top_actors.index, hue=top_actors.index, palette="turbo", legend=False)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)
plt.xlabel('Revenue(Milions)', fontsize=12)
plt.ylabel('Directors', fontsize=12)
plt.title("Top 10 Directors Who Made the Most Moneys", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Splitting the 'Genre' column by commas to create a list of genres for each movie
df["Genre"] = df['Genre'].str.split(",")

# Exploding the list of genres so that each genre is in a separate row
df1 = df.explode("Genre")
```

```python
# Counting the occurrences of each genre in the 'Genre' column after exploding
genres = df1['Genre'].value_counts()

# Displaying the genre counts
genres
```

```python
# Set the style of the plot to white grid
sns.set_style("whitegrid")

# Create a figure with specific size
plt.figure(figsize=(10,6))

# Create a bar plot with genres and their corresponding values
ax = sns.barplot(   
                    x=genres.values,  # Movie count for each genre
                    y=genres.index,   # Genre names
                    hue=genres.index,  # Color bars based on genres
                    palette="viridis",  # Set the color palette
                    legend=False)  # Disable the legend

# Add labels to each bar with the exact number on top
for container in ax.containers:
    ax.bar_label(   
                    container,  # The container of bars
                    fmt='%d',  # Format labels as integers
                    label_type='edge',  # Place labels at the edge
                    fontsize=12,  # Set font size for labels
                    color='black',  # Set label color to black
                    padding=3)  # Add padding to label position

# Set the labels for X and Y axes
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Genres', fontsize=12)

# Set the title for the plot
plt.title("Movie Genres", fontsize=14, fontweight='bold')

# Show the plot
plt.show()
```

```python
genres_norm = (df1['Genre'].value_counts(normalize=True) * 100)[:10]
genres_norm
```

```python
# Calculate the normalized value counts (percentage) for the top 10 genres
genres_norm = (df1['Genre'].value_counts(normalize=True) * 100)[:10]

# Create a figure and axis with a specific size
fig, ax = plt.subplots(figsize=(8, 8))

# Create a pie chart for the top 10 genres
ax.pie(genres_norm.values,  # Values to plot (percentages)
       labels=genres_norm.index,  # Labels for the genres
       autopct='%1.1f%%',  # Format the percentage with one decimal place
       startangle=90,  # Start angle for the first slice (rotating the chart)
       textprops={'fontsize': 10})  # Set the font size for the labels

# Set the title for the pie chart
ax.set_title('The 10 Most Filmed Genres', fontsize=16, fontweight='bold')

# Equalize the axes to ensure the pie chart is circular
ax.axis('equal')

# Show the pie chart
plt.show()
```
